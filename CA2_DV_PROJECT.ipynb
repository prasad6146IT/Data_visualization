```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_data=pd.read_csv('/content/covid19datasetprasad.csv')
df_data.head()
```

```python
correlation_matrix=df_data.corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title("correlation heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



plt.figure(figsize=(8,6))
sns.scatterplot(x='Recovered', y='Deaths',hue='WHO Region ',data=df_data)
plt.title("Recovered vs Deaths")
plt.xlabel("Recovered case")
plt.ylabel("Deaths case")

plt.show()

```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.countplot(x='WHO Region ',data=df_data)
plt.title("WHO Regions")

plt.xticks(rotation=45)
plt.show()
```

```python
df_total=df_data[['Confirmed','Recovered','Deaths']].sum()
label=['Confirmed','Recovered','Deaths']
color=['aqua','purple','orange']
plt.figure(figsize=(8,6))
plt.pie(df_total,labels=label,autopct='%1.1f%%',explode=(0,0.1,0),colors=color,startangle=140,shadow=True)
plt.title('DISTRIBUTION OF ALL CASES WORLDWIDE',fontsize=10)
plt.axis('equal')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
region_total=df_data.groupby('Region')['Confirmed'].sum().nlargest(7)

plt.figure(figsize=(8,6))
plt.pie(region_total,labels=region_total.index,autopct='%1.1f%%',startangle=140,shadow=True)
plt.title('DISTRIBUTION OF CONFIRMED CASES BY COUNTRY',fontsize=10,color='r',loc='left',horizontalalignment='center')
plt.axis('equal')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Group data by WHO Region and calculate the sum of confirmed cases and deaths
region_data = df_data.groupby('WHO Region ')[['Confirmed', 'Deaths']].sum()
plt.figure(figsize=(10, 6))
plt.plot(region_data['Confirmed'], label='Confirmed Cases', marker='o')
plt.plot(region_data['Deaths'], label='Deaths', marker='x')
plt.xlabel('WHO Region')
plt.ylabel('Count')
plt.title('Confirmed Cases vs. Deaths by WHO Region')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
